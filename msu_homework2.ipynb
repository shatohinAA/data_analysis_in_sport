```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
chunk = pd.read_csv('df_events_epl_train.csv',chunksize=50000)
df = pd.concat(chunk)
```

```python
df
```

```python
passes = df[df.type_name=='Pass']
passes
```

```python
passes.drop(passes[(passes.pass_outcome_name == 'Pass Offside') | (passes.pass_outcome_name == 'Unknown') | (passes.pass_outcome_name == 'Injury Clearance') | (passes.pass_type_name == 'Kick Off') | (passes.pass_type_name == 'Throw-in') | (passes.pass_type_name == 'Goal Kick') | (passes.pass_type_name == 'Corner') | (passes.pass_type_name == 'Free Kick')].index, inplace=True)
```

```python
passes
```

```python
passes = passes.dropna(axis=1,how='all')
passes
```

```python
len(passes[passes.outcome_binary==0].index)
```

```python
y = passes.outcome_binary
X = passes[['period','minute','second','possession','possession_team_id','play_pattern_id','team_id','duration','player_id','position_id','location_0','location_1','pass_length','pass_angle','pass_height_id','pass_end_location_0','pass_end_location_1','pass_type_id','pass_body_part_id','under_pressure','pass_technique_id']]
```

```python
X = X.fillna(value={'pass_type_id': 1, 'pass_body_part_id': 1, 'under_pressure': 0, 'pass_technique_id': 1
                   })
X = X.replace({'under_pressure': True}, 1)
```

Далее пробуем различные модели и подбираем их параметры.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, BaggingClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold, GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)#, random_state=42)
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
rfc = RandomForestClassifier(random_state=42, n_jobs=-1, oob_score=True)
results = cross_val_score(rfc, X_train, y_train, cv=skf)

print("CV accuracy score: {:.2f}%".format(results.mean()*100))
results
```

```python
rfc.fit(X_train, y_train)
predictions = rfc.predict_proba(X_test)
auc = roc_auc_score(y_test.values.reshape(-1,1), predictions[:,1].reshape(-1,1))
auc
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.astype(np.float64))
X_test_scaled = scaler.fit_transform(X_test.astype(np.float64))

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
rfc_scaled = RandomForestClassifier(random_state=42, n_jobs=-1, oob_score=True)
results = cross_val_score(rfc_scaled, X_train_scaled, y_train, cv=skf)

print("CV accuracy score: {:.2f}%".format(results.mean()*100))
results
```

```python
rfc_scaled.fit(X_train_scaled, y_train)
predictions = rfc_scaled.predict_proba(X_test_scaled)
auc = roc_auc_score(y_test.values.reshape(-1,1), predictions[:,1].reshape(-1,1))
auc
```
